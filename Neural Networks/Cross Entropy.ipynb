# Calculating Cross Entropy

Now, time to shine! Let's code the formula for cross-entropy in Python. As in the video, **Y** in the quiz is for the category, and **P** is the probability. 

For the picture below, think of each door as a feature.  And you are looking at a row, and calculateing the cross entropy.

```python
display.Image('./images/cross-entropies.png')
```

```python
def cross_entropy(Y, P):
    Y = np.float_(Y)
    P = np.float_(P)
    return -np.sum(Y * np.log(P) + (1 - Y) * np.log(1 - P))
```

```python
# Test
Y=[1,0,1,1]
P=[0.4,0.6,0.1,0.5]

cross_entropy(Y,P)
```
